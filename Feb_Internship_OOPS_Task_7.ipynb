```python
'''
Problem 1: Smart Light Controller
Problem Statement
Create a class to control a smart light that can be turned ON or OFF and display its status.
🌍 Real-Time Use
Smart home automation systems
📥 Sample Input
Light Name: Bedroom Light
Action: ON
📤 Expected Output
Bedroom Light is ON

'''

class SmarLightController:
    def __init__(self,light_name):
        self.light_name = light_name
        self.state="off"
    def status(self):
      if self.state=="on":
        print(f"{self.light_name} is {self.state}")
      else:
        print(f"{self.light_name} is {self.state}")
    def update_state(self,state):
      self.state=state
      self.status()
flag=True;
light = None # Initialize light object

print("Menu: ")
print("Enter 1 to turn on the light")
print("Enter 2 to turn off the light")
print("Enter 3 to check the status of the light")
print("Enter 4 to change light name")
print("Enter 5 to exit")

while(flag):
  choice=int(input("Enter your choice: "))
  if choice==1:
    if light is None:
      light_name = input("Enter the light name: ")
      light = SmarLightController(light_name)
    light.update_state("on")
  elif choice == 2:
      if light:
          light.update_state("off")
      else:
          print("No light created yet. Please create a light first (Option 1).")
  elif choice == 3:
      if light:
          light.status()
      else:
          print("No light created yet. Please create a light first (Option 1).")
  elif choice == 4:
      if light:
          new_name = input("Enter the new light name: ")
          light.light_name = new_name
          print(f"Light name changed to {new_name}")
      else:
          print("No light created yet. Please create a light first (Option 1).")
  elif choice == 5:
      flag = False
      print("Exiting Smart Light Controller.")
  else:
      print("Invalid choice. Please enter a number between 1 and 5.")

```

```python
'''
Problem 4: Product Price Tag Generator
Problem Statement
Create a class that stores product name and price and prints a formatted price tag.
 Real-Time Use
Retail billing systems
📥 Sample Input
Product Name: Headphones
Price: 2499
📤 Expected Output
Product: Headphones
Price: ₹2499
'''

class Product:
  def __init__(self):
    self.name=input("enter prod name: ")
    self.price=float(input("enter price "))
  def show(self):
    print("Product: ",self.name)
    print(f"Price: ₹{self.price:.2f}")

prod=Product()
prod.show()
```

```python
'''
Problem 5: Movie Rating Display System
Problem Statement
Create a class that stores movie name and rating and displays movie details.
Real-Time Use
Streaming platforms
📥 Sample Input
Movie Name: Inception
Rating: 4.8

'''
class Movies:
    ratings = {}

    def __init__(self):
        self.name = ""
        self.rating = 0.0

    def add(self):
        self.name = input("Enter movie name: ").strip().upper()
        self.rating = float(input("Enter rating out of 5: "))

        if self.name in Movies.ratings:
            Movies.ratings[self.name]["total"] += self.rating
            Movies.ratings[self.name]["count"] += 1
        else:
            Movies.ratings[self.name] = {"total": self.rating, "count": 1}

        print("Rating added successfully.")

    def get_rating(self):
        self.name = input("Enter movie name: ").upper()

        if self.name in Movies.ratings:
            total = Movies.ratings[self.name]["total"]
            count = Movies.ratings[self.name]["count"]
            avg = total / count
            print(f"Movie: {self.name}\nAverage Rating: {avg:.2f}")
        else:
            print("No data found.")

    def show(self):
        if not Movies.ratings:
            print("No movies available.")
            return

        print("\nMovie Info:")
        for name, data in Movies.ratings.items():
            avg = data["total"] / data["count"]
            print(f"Movie: {name}\nAverage Rating: {avg:.2f}\n")


flag = True
mov = Movies()

print("Menu")
print("1 - Add movie rating")
print("2 - Get movie rating")
print("3 - Show all movies")
print("4 - Exit")

while flag:
    choice = int(input("Enter your choice: "))

    if choice == 1:
        mov.add()
    elif choice == 2:
        mov.get_rating()
    elif choice == 3:
        mov.show()
    elif choice == 4:
        flag = False
    else:
        print("Invalid choice")
```
